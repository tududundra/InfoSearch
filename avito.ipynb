```python
def get_soup(url):
    try:
        req = requests.get(url)
        assert req.status_code == 200, 'request failed'
    except AssertionError:
        bad.append(url)
    
    html = req.text
    soup = BeautifulSoup(html, 'lxml')
    return file_maker(url, soup)
```

```python
def file_maker(url, soup):
    url_name = str(re.sub('[/&?~#%\*\}\{\\:<>\|\"]', '_', url))[8:]
    print(url_name)
    file_name = url_name + '.txt'
    #print(file_name)
    
    if not os.path.isfile(file_name):
    
        with open(file_name, 'w', encoding='utf-8') as flilo:
            name = soup.find('span', {'class': 'title-info-title-text'})
            if name is None:
                bad.append(url)
            else:

                name = name.text



                data = soup.find('div', {'class': 'title-info-metadata-item'})
                if data is None:
                    data = ''
                else:
                    data = data.text


                price = soup.find('span', {'class': 'js-item-price'})
                if price is None:
                    price = soup.find('span', {'class': 'price-value-string js-price-value-string'})

                price = price.text

                text_ = soup.find('div', {'class': 'item-description-text'})
                if text_ is None:
                    text_ = soup.find('div', {'class': 'item-description-html'})

                if not text_ is None:
                    text_ = text_.text
                    full = str(name) + '\n' + str(price) + '\n' + str(data) + '\n' + str(text_) + '\n'
                    #print(full)
                    flilo.write(full)

                else:
                    bad.append(str(url))
                #full = str(name) + '\n' + str(price) + '\n' + str(data) + '\n' + str(text_) + '\n'
                #print(full)
                #flilo.write(full)
        
```
